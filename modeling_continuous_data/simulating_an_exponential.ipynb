```python
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
import numpy as np

rng = np.random.default_rng(2021)

font_size = 25
font_size_legend = 20
font_size_ticks = 20

n = 10000
u = rng.random(n)
# print(u)
n_bins = 50

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(u,bins=n_bins,color='white', edgecolor='black', linewidth=2,
         density=True)
# plt.ylim([0,y_max])
plt.savefig('plots/uniform_samples.pdf',bbox_inches="tight")

transformed_u = np.log(1/(1-u))
lambda_ml = 1/np.mean(transformed_u)
print(lambda_ml)

step = 1e-2
t=np.arange(0,np.max(transformed_u)+step,step)
exponential_fit = expon.pdf(t,0,1/lambda_ml)

y_max = 1.01

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(transformed_u,bins=n_bins,color='white', edgecolor='black', linewidth=2,
         density=True)
plt.ylim([0,y_max])
plt.savefig('plots/transformed_uniform_samples.pdf',bbox_inches="tight")

plt.figure(figsize=(9,9))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.hist(transformed_u,bins=n_bins,color='white', edgecolor='dimgray', linewidth=2,
         density=True)
plt.plot(t,exponential_fit,lw=2,color="black",
     label=r'Parametric model $f_{\lambda}(t)$')
plt.ylim([0,y_max])
plt.savefig('plots/transformed_uniform_samples_exponential.pdf',bbox_inches="tight")

```
